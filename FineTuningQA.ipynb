```python
embedding_model = SentenceTransformer('multi-qa-mpnet-base-dot-v1')
```

```python
def embedding_data(context):
    try:
        if not isinstance(context, str):
            raise ValueError("Input context must be a string.")
        embedding = embedding_model.encode(context)
        return embedding
    except Exception as e:
        print(f"An error occurred during embedding generation: {e}")
        return None
```

```python
def get_answer(query):
    try:
        response = Groq.chat.completions.create(
            messages=[
                {"role": "system", "content": "You are a helpful assistant."},
                {"role": "user", "content": query}
            ],
            max_tokens=150
        )
        return response["choices"][0]["message"]["content"]
    except Exception as e:
        print(f"An error occurred during answer generation: {e}")
        return None
```

```python
train_df['questions'] = train_df['questions'].astype(str)
train_df['Embedding_Q'] = train_df.questions.apply(embedding_data)
```

```python
train_df['answer'] = train_df['answer'].astype(str)
train_df['Embedding_A'] = train_df.answer.apply(embedding_data)
```

```python
from scipy.spatial.distance import cosine
import pandas as pd
import numpy as np

def rank_distances(questions_embedding, answer_embeddings_list):
    distances = []
    for i, embedding in enumerate(answer_embeddings_list):
        if isinstance(embedding, str):
            try:
                embedding = eval(embedding)
            except (ValueError, SyntaxError):
                print(f"Invalid embedding format at index {i}: {embedding}")
                continue

        if isinstance(embedding, list):
            embedding = np.array(embedding, dtype=float)

        distance = cosine(questions_embedding, embedding)
        distances.append((i, distance))

    return sorted(distances, key=lambda x: x[1])

question_embedding = train_df["Embedding_Q"].iloc[0]

ranked_results = rank_distances(question_embedding, train_df["Embedding_A"])

print("Pertanyaan: ", train_df["questions"].iloc[0])
top_5_results = [
    {"No": idx + 1, "Rank": round(rank, 6), "Answer": train_df["answer"].iloc[i]}
    for idx, (i, rank) in enumerate(ranked_results[:5])
]

top_5_df = pd.DataFrame(top_5_results)
print(top_5_df)
```
